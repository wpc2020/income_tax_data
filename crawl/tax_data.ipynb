```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
county_dict={'A':'臺北市','B':'臺中市','C':'基隆市','D':'臺南市','E':'高雄市','F':'新北市','G':'宜蘭縣','H':'桃園市','I':'嘉義市','J':'新竹縣','K':'苗栗縣','L':'臺中縣','M':'南投縣','N':'彰化縣','O':'新竹市','P':'雲林縣','Q':'嘉義縣','R':'臺南縣','S':'高雄縣','T':'屏東縣', 'U':'花蓮縣', 'V':'臺東縣','W':'金門縣','X':'澎湖縣','Z':'連江縣'} 
```

```python
# 總額分析
# 把資料抓下來

def get_html(letter):
    url = url_core +letter+'.html'
    r = requests.get(url)
    html = r.text
    html = html.encode('ISO-8859-1')
    html = html.decode('utf-8')
    return html

def get_df(html, letter):
    soup = BeautifulSoup(html, 'lxml')
    # 創造空的dataframe
    df = pd.DataFrame(columns=['縣市','鄉鎮市區', '村里', '納稅單位','綜合所得總額', '平均數','中位數','第一分位數','第三分位數','標準差','變異係數'])
    # 把縣市名稱補上去
    # matches = 要刪除的東西
    county=county_dict[letter]
    matches = ['綜稅所得總額各縣市鄉鎮村里統計分析表','單位：金額(千元)','製表日期', '縣市別','本統計']
    # 該網頁用兩層table包住
    table = soup.find_all('table')[0]

    for table in table.find_all('table'):
        for row in table.find_all('tr'):
            col_num = len(row.find_all('span'))
            # 第一欄是鄉鎮市區
            if col_num ==10:
                new=[county,]
                town=row.find_all('span')[0].text.strip().replace('\u3000','').replace(u'\xa0','')
                for col in row.find_all('span'):
                    new.append(col.text.strip().replace('\u3000','').replace(u'\xa0',''))
            # 第一欄是村里
            elif col_num ==9:
                new=[county,town,]
                for col in row.find_all('span'):
                    new.append(col.text.strip().replace('\u3000','').replace(u'\xa0','') ) 
            # 合併回去原本的dataframe
            if col_num ==10 or col_num ==9 :
                df.loc[len(df)] = new
    return df
```

```python
# 各類所得
# 把資料抓下來
def get_water(letter):
    url = 'https://ws.fia.gov.tw//001/Upload/public/attachment/isa100/100_166-'+letter+'.html'
    r = requests.get(url)
    html = r.text
    html = html.encode('ISO-8859-1')
    html = html.decode('utf-8')
    return html

def get_soup(html, letter):
    soup = BeautifulSoup(html, 'lxml')
    # 創造空的dataframe
    df = pd.DataFrame(columns=['縣市','鄉鎮市區', '村里', '納稅單位','合計','營利所得','執行業務所得','薪資所得','利息所得','租賃及權利金','財產交易所得','機會中獎所得'])
    # 把縣市名稱補上去
    county_dict={'A':'臺北市','B':'臺中市','C':'基隆市','D':'臺南市','E':'高雄市','F':'新北市','G':'宜蘭縣','H':'桃園市','I':'嘉義市','J':'新竹縣','K':'苗栗縣','L':'臺中縣','M':'南投縣','N':'彰化縣','O':'新竹市','P':'雲林縣','Q':'嘉義縣','R':'臺南縣','S':'高雄縣','T':'屏東縣', 'U':'花蓮縣', 'V':'臺東縣','W':'金門縣','X':'澎湖縣','Z':'連江縣'}
    # matches = 要刪除的東西
    county=county_dict[letter]
    matches = ['各類所得金額各縣市鄉鎮村里統計表','單位：金額(千元)','製表日期', '縣市別','本統計']
    # 該網頁用兩層table包住
    table = soup.find_all('table')[0]

    for table in table.find_all('table'):
        for row in table.find_all('tr'):
            col_num = len(row.find_all('span'))
            # 第一欄是鄉鎮市區
            if col_num ==11:
                new=[county,]
                town=row.find_all('span')[0].text
                for col in row.find_all('span'):
                    new.append(col.text)
            # 第一欄是村里
            elif col_num ==10:
                new=[county,town,]
                for col in row.find_all('span'):
                    new.append(col.text) 
            # 合併回去原本的dataframe
            if col_num ==11 or col_num ==10 :
                df.loc[len(df)] = new
    return df
```

```python
# 土法煉鋼減少append運算負擔
letters = ['B','C','D','E',]
df1 = get_soup(get_water('A'),'A')
for letter in letters:
    html=get_water(letter)
    if html!='':
        temp=get_soup(html,letter)
        df1=df1.append(temp)
        
letters = ['G','H','I','J','K','L']
df2 = get_soup(get_water('F'),'F')
for letter in letters:
    html=get_water(letter)
    if html!='':
        temp=get_soup(html,letter)
        df2=df2.append(temp)        
        
letters = ['N','O','P','Q','R','S','T','U','V','W','X','Z']
df3 = get_soup(get_water('M'),'M')
for letter in letters:
    html=get_water(letter)
    if html!='':
        temp=get_soup(html,letter)
        df3=df3.append(temp)
df=pd.concat([df1,df2,df3],ignore_index=True)
```
